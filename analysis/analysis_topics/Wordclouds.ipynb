```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud as WC

import csv
```

```python
def is_even(n: int):
    return n % 2 == 0

topics = dict()

with open("topic_description.csv", encoding = "utf-8-sig") as f:
    reader = csv.reader(f, dialect='excel')
    pair = list()
    try:
        for i,line in enumerate(reader):
        
            # if we reach an even i, we have collected one list of words
            # and one list of weights. add into topics dictionary
            if is_even(i) and i > 0:
                # topic is the first entry in the words list
                topic = pair[0][0]
                # the value is a dictionary with words and their weights
                topics[topic] = dict(zip(pair[0][1:], [float(weight) for weight in pair[1][1:]]))
                # finally, empty the list
                pair = list()
                # append new line
                pair.append(line)
            else:
                pair.append(line)
        # add last topic to dictionary
        topic = pair[0][0]
        topics[topic] = dict(zip(pair[0][1:], [float(weight) for weight in pair[1][1:]]))
    except UnicodeDecodeError as e:
        print(e)
        print(i, line)
```

```python
for topic_name in topics.keys():
    print(topic_name)
    # get the dictionary of words and their weights
    topic = topics[topic_name]
    # restrict to the more relevant words
    #topic = {k:v for k,v in topic.items() if v > 0.009}
    
    # initialize wordcloud class
    wc = WC(background_color='white', width=500, height=300, margin=2, prefer_horizontal=1)
    wc.fit_words(topic)
    wc.recolor(color_func=lambda *args, **kwargs: (0,0,0))
    wc.to_file(f'WordClouds/{topic_name}.png')
```
