In this notebook, we generate a plot of perplexity scores obtained from the training dataset.

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rc

rc('text', usetex=True)

perplexity_values = np.loadtxt("perplexity.csv", delimiter=",")

# Create array for x values
x_values = np.array(range(1, 151))

# Create a new figure
plt.figure(figsize=(10,6))

# Plot perplexity values
plt.plot(x_values, perplexity_values)

# Add vertical line at 70
plt.axvline(x=70, color='r', linestyle='--')

# Add labels and title
plt.xlabel('Samples')
plt.ylabel('Perplexity')

plt.tight_layout()

# Save as .eps file
plt.savefig('perplexity_plot.eps', format='eps')

# Show the plot
plt.show()
```
