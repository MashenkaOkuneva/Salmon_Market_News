```python
import pandas as pd
import os

path = os.path.realpath(__name__) # path
drt = os.path.dirname(path)       # directory

# Load the CSV file into a DataFrame
file_path = drt.replace('\\robustness_check', '') + '\\analysis\\VAR' + '/daily_topics_extend_comp_rev_components.csv'
daily_topics = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
daily_topics.head()
```

```python
# Load the Excel file into a DataFrame
file_path = 'OSE_Prices.xlsx'
ose = pd.read_excel(file_path)

# Display the first few rows of the DataFrame
ose.head()
```

```python
# Ensure that both 'date' columns are in datetime format
daily_topics['date'] = pd.to_datetime(daily_topics['date'])
ose['Exchange Date'] = pd.to_datetime(ose['Exchange Date'])

# Merge the DataFrames on the date columns
# Rename the 'Exchange Date' column to 'date' in the ose DataFrame for merging
ose.rename(columns={'Exchange Date': 'date'}, inplace=True)

# Merge the DataFrames
merged_df = pd.merge(daily_topics, ose[['date', 'log_ret_OSE']], on='date', how='left')

# Display the first few rows of the merged DataFrame
merged_df.head()
```

```python
# Save the DF:
save_path = drt.replace('\\robustness_check', '') + '\\analysis\\VAR' + '/daily_topics_extend_comp_rev_components_ose.csv' 
merged_df.to_csv(save_path, index = False)
```
